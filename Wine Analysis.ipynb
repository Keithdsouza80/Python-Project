# **4. Data Acquisition & Description**
---

|Id|Feature|Description|
|:--|:--|:--|
|01|fixed acidity| most acids involved with wine are fixed or nonvolatile (do not evaporate readily).| 
|02|volatile acidity|the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste.| 
|03|citric acid|found in small quantities, citric acid can add ‘freshness’ and flavor to wines.| 
|04|residual sugar | the amount of sugar remaining after fermentation stops, it’s rare to find wines with less than 1 gram/liter |
|||and wines with greater than 45 grams/liter are considered sweet.|
|05| chlorides| the amount of salt in the wine.|
|06| free sulfur dioxide  | free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion;|
|||it prevents microbial growth and the oxidation of wine.|
|07|  total sulfur dioxide | amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine,|
|||but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine|
|08| density| the density of water is close to that of water depending on the percentage of alcohol and sugar|
|09| pH | describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic);|
||| most wines are between 3-4 on the pH scale.|
|10| sulphates| a wine additive for preservation which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant.|
|11| alcohol| the percentage of alcohol content of the wine.|
|12| quality| score between 0 to 10.|

#### Analysing Volatile Acidity
- We see that volatile acidity is almost equally spread across the wine samples.

- However, there are a few wines which are highly acidic crossing the 1.4 gms/ltr normally associated with Red wines

```python
wq['volatile acidity'].plot(kind='area', figsize=(15,7), color='violet', fontsize=13)

plt.xlabel('Index')
plt.ylabel('Grams/Litre')
plt.title('Line chart showing the variation of volatile acidity across wines')
```

---
<a name = Section6></a>
#### Analysing Quality
- 93.2% of the data is of 5 to 7 grade quality, only 3.1% is above this Quality 7% 

```python
space = np.ones(30)/20
wq ['quality'].value_counts().plot(kind='pie', fontsize=14, startangle=150,figsize=(25,25), autopct='%3.1f%%',
                                   cmap='cool', legend=True)
plt.ylabel('')
plt.title('Pie Chart showing the quality rating')
```

- Observing more closely on this Histogram on the Quality. We come to the same conclusion that we are analysing average 
  to more than average quality wines   

```python
wq ['quality'].plot(kind='hist', figsize=(10, 7), color='brown', yticks=np.arange(10, 3))

plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.title('Histogram showing the distribution of quality')
```

# **7. Exploratory Data Analysis**

---
<a name = Section7></a>
#### Analysing Volatile acidity and Citric acid
- Volatile acidity is between 0 to 0.5 which is a sign of a good quality wine (below 1.2 being good for whites and below 1.4 being good for reds)
- In the previous observation we saw a few outliers which can be ignored.  
- Lower concertration of Citric acid is also a sign of a good wine

```python
wq['volatile acidity'].plot(kind='kde', figsize=(15, 7), color='black')

plt.xlabel('volatile acidity')
plt.ylabel('quantity')
plt.title('KDE plot showing the distribution of volatile acidity')
```

```python
wq['citric acid'].plot(kind='kde', figsize=(15, 7), color='black')

plt.xlabel('citric acid')
plt.ylabel('quantity')
plt.title('KDE plot showing the quantity of citric acid')
```

---
<a name = Section7></a>
#### Analysing Sulfur dioxide
- Free and Total Sulfur dioxide are normally in line with each other. If Free SO2 is less in then Total SO2 i released to balance out the Sulfur content
- From the Stacked Area Chart below, it can be seen that both are balance concluding that the wines are good quality
- Additionally over 50 for free SO2 is seems to be found in the lower quality wine, concluding further that these are mostly good wines

```python
wq.groupby(['quality'])['total sulfur dioxide'].mean().plot(kind='area', figsize=(15, 7), color='#347593')
wq.groupby(['quality'])['free sulfur dioxide'].mean().plot(kind='area', figsize=(15, 7), color='#363552', fontsize=13)
plt.xlabel('quality')
plt.ylabel('sulfur')
plt.title('Stacked Area Chart showing the free and total sulfur by wine quality')
plt.legend(labels=('total sulfur dioxide', 'free sulfur dioxide'))
```

---
<a name = Section7></a>
#### Analysing pH
- pH proves the acidity or blandness of the wine. All the wines are between the 3 - 4 range which is a good sign.
- However, one can see that the range varies depending on the quality.

```python
wq.groupby(['quality'])['pH'].mean().plot(kind='line', figsize=(15, 7), color='#347593', fontsize=13)
plt.xlabel('quality')
plt.ylabel('pH')
plt.title('Stacked Area Chart showing the pH by wine quality')
```

---
<a name = Section7></a>
#### Analysing correlation
- There is a strong positive correlation between both fixed and free SO2
- However, there is a negative correlation between Alcohol and Water and positive correlation between Alcohol and Quality.
  Concluding that higher the Alcohol and lower the water content better the wine
- Also, higher the water (density) higher the residual sugar thus decreasing the quality of the Wine.

```python
corr_mat = wq.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_mat, annot=True, cmap='viridis')
```
